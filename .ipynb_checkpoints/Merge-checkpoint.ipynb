```python
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

def SourceCombination(country,tweet_path, oxcgrt_path, oxcgrt_va_path):
    """
    Function to merget tweet, OxCGRT and OxCGRT Vacciantion data
    :param: country: contry name
    :param: tweet_path to tweet data
    :param: oxcgrt_path to OxCGRT_latest.csv
    :param: oxcgrt_path to OxCGRT_vaccination_full.csv
    """
    ################## read tweet file
    df = pd.read_csv(tweet_path)
    start_date = dt.datetime.strptime('2020-01-1','%Y-%m-%d')
    df['Date'] = df.time.apply(lambda x: dt.datetime.strptime(x.split(' ')[0],'%Y-%m-%d'))
    
    ################## build tweet index: get average likes or retweet in each 7 days period
    likes_index = []
    retweet_index = []
    date_index = []
    for i in range(int(623/7)): 
        df_temp = df[(df['Date']>=start_date + i*7*dt.timedelta(days=1)) & (df['Date']<start_date + (i+1)*7*dt.timedelta(days=1)) ]
        date_index = date_index + [start_date + i*7*dt.timedelta(days=1) + dt.timedelta(days=k) for k in range(7)]
        likes_index = likes_index + [df_temp.likes.mean()]*7
        retweet_index = retweet_index + [df_temp.retweets.mean()]*7
    df_like_index = pd.DataFrame({'Date':date_index, 'like_index':likes_index, 'retweet_index': retweet_index})
    df_like_index = df_like_index.fillna(0)
    
    ################## read OxCGRT file
    df_ocgrt = pd.read_csv(oxcgrt_path)
    df_ocgrt = df_ocgrt[df_ocgrt['Jurisdiction'] == 'NAT_TOTAL']
    df_Ox_country = df_ocgrt[df_ocgrt.CountryName== country]
    df_Ox_country['Date'] = df_Ox_country['Date'].apply(lambda x: dt.datetime.strptime(str(x),'%Y%m%d'))
    df_Ox_country['Daily_cases'] = df_Ox_country['ConfirmedCases'].diff()
    
    ################## Merge tweet and OxCGRT
    df_Ox_country_tweets = df_Ox_country.merge(df_like_index,'left', on = 'Date')
    
    ########### Chart
    df_Ox_country_tweets.plot(x='Date',y=['like_index','retweet_index'],figsize=(10,4))
    ax2 = plt.twinx()
    sns.lineplot(x =df_Ox_country_tweets.Date, y = df_Ox_country_tweets.Daily_cases, ax = ax2,linewidth=.4,color='red')
    plt.savefig('Picture/Preliminary_plot_{}.png'.format(country))
    
    
    df_va =pd.read_csv(oxcgrt_va_path)
    df_va = df_va[df_va.CountryName == country]
    df_va['Date'] = df_va['Date'].apply(lambda x: dt.datetime.strptime(str(x),'%Y%m%d'))
    df_Ox_country_tweets_va = df_Ox_country_tweets.merge(df_va,'left',on='Date')
    
    df_Ox_country_tweets_va.to_csv('Data/Full_{}.csv'.format(country))
    
    
    with open('Data/columns.txt','w') as f:
        for column in df_Ox_country_tweets_va.columns.to_list():
            f.write(column + '\n')
    f.close()
```

```python
# Setting 
oxcgrt_path = 'Data/covid-policy-tracker-master/data/OxCGRT_latest.csv'
oxcgrt_va_path = 'Data/covid-policy-tracker-master/data/OxCGRT_vaccines_full.csv'
countries = ['Brazil', 'India','United States','Mexico']
tweet_file = ['Finalized_Brazil.csv','Finalized_India.csv','Finalized_USA.csv','Finalized_Mexico.csv']
```

```python
# Brazil
SourceCombination(countries[0],'Data/Tweets/'+tweet_file[0], oxcgrt_path, oxcgrt_va_path)
```
