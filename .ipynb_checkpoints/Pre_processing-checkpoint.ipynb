```python
import os
import re
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
def init_db(countries, datadir):
    
    if not os.path.isdir(datadir):
        os.mkdir(datadir)

    dbs = []*len(countries)
    for i in range(len(countries)):
        datafile = "Finalized_" + countries[i] + ".csv"
        
        if os.path.isfile(os.path.join(datadir, datafile)):
            # if old file exists, get the db
            #dbs.append(pd.read_csv(os.path.join(datadir, datafile), index_col=0))
            os.remove(os.path.join(datadir, datafile))
        dbs.append(pd.DataFrame(columns=['username', 'retweets', 'likes', 'text', 'hashtags', 'time']))
        
    return dbs
```

```python
def fetch_data(dbs, datadir, countries):
    
    [US, Mexico, Brazil, India] = dbs
    
    for country in countries:
        print("##### Reading", country, "#####")
        savingdir = os.path.join(datadir, country)
        for folder in os.scandir(savingdir):
            for f in os.scandir(folder):
                filepath = os.path.join(savingdir, folder.name, f.name)
                data = pd.read_csv(filepath, index_col=0)
                if country == 'United States':
                    US = US.append(data, ignore_index=True)
                elif country == 'Mexico':
                    Mexico= Mexico.append(data, ignore_index=True)
                elif country == 'Brazil':
                    Brazil= Brazil.append(data, ignore_index=True)
                elif country == 'India':
                    India = India.append(data, ignore_index=True)
                else:
                    print(f.name)
                    print("Error")
        updated = [US, Mexico, Brazil, India]
    return updated 
```

```python
def tweets_combination(datadir, dbs):
    
    ## Criteria for COVID-19 data
    keywords = ['ncov', 'corona', 'virus', 'covid', 'sars-cov-2', 'unknown pneumonia', 'corona', '2019-nCov', 'COVID-19', 'DeltaVariant', 'workfromhome', 'WFH', 'social distance', 'stayhome', 'gatherings restrictions', 'lockdown', 'quarantine', 'reopen', 'immunization', 'vaccine', 'International support', 'debt relief', 'fiscal measures', 'GDP', 'recession', 'testing', 'mask mandate', 'mask', 'vaccine development', 'protect elderly', 
               'कोरोनावाइरस','अज्ञात निमोनिया','कोरोना','डेल्टा', 'IndiaFightsCorona',
               'vírus', 'pneumonia desconhecida', 'Vacina','vacinações','confinamento','quarentena','fecha','fecho', 'Vacinação', 'imunização', 'Suporte internacional', 'alívio da dívida', 'medidas fiscais', 'PIB', 'mascarar','proteger','idosos','traçando', 'testando', 'estímulo', 'PCR', 'recessão',
               'neumonía desconocida', 'Vacunas','trabajar desde casa','Quédate en casa','distancia social','cuarentena','cerrar','cierre','inmunización','Apoyo internacional','deuda','medidas fiscales','PIB','recesión','máscara','mandato de máscara','desarrollo de vacunas','proteger a los ancianos','vacunar','estímulo','ingreso','exportar','desempleo','pruebas','Rastreo']
    
    tmp_dbs = []
    for db in dbs:
        tmp_db = db.copy()
        for i in range(len(db)):
            contains = False
            for word in keywords:
                # Filter keywords
                if word in tmp_db.loc[i,'text'] or word in tmp_db.loc[i,'hashtags']:
                    contains = True
                    break
                    
            if not contains:
                tmp_db.drop(i, axis=0, inplace=True)

        tmp_dbs.append(tmp_db)
    [US, Mexico, Brazil, India] = tmp_dbs 

    US.to_csv(os.path.join(datadir, 'Finalized_United States.csv'))
    Mexico.to_csv(os.path.join(datadir, 'Finalized_Mexico.csv'))
    Brazil.to_csv(os.path.join(datadir, 'Finalized_Brazil.csv'))
    India.to_csv(os.path.join(datadir, 'Finalized_India.csv'))
    return [US, Mexico, Brazil, India]
```

```python
def source_combination(countries, tweet_path, oxcgrt_path, oxcgrt_va_path):
    """
    Function to merget tweet, OxCGRT and OxCGRT Vacciantion data
    :param: country: contry name
    :param: tweet_path to tweet data
    :param: oxcgrt_path to OxCGRT_latest.csv
    :param: oxcgrt_path to OxCGRT_vaccination_full.csv
    """
    ################## read tweet file
    warnings.filterwarnings("ignore")
    df = pd.read_csv(tweet_path)
    start_date = dt.datetime.strptime('2020-01-1','%Y-%m-%d')
    df['Date'] = df.time.apply(lambda x: dt.datetime.strptime(x.split(' ')[0],'%Y-%m-%d'))
    
    ################## build tweet index: get average likes or retweet in each 7 days period
    likes_index = []
    retweet_index = []
    date_index = []
    for i in range(int(623/7)): 
        df_temp = df[(df['Date']>=start_date + i*7*dt.timedelta(days=1)) & (df['Date']<start_date + (i+1)*7*dt.timedelta(days=1)) ]
        date_index = date_index + [start_date + i*7*dt.timedelta(days=1) + dt.timedelta(days=k) for k in range(7)]
        likes_index = likes_index + [df_temp.likes.mean()]*7
        retweet_index = retweet_index + [df_temp.retweets.mean()]*7
    df_like_index = pd.DataFrame({'Date':date_index, 'like_index':likes_index, 'retweet_index': retweet_index})
    df_like_index = df_like_index.fillna(0)
    
    ################## Read OxCGRT file
    df_ocgrt = pd.read_csv(oxcgrt_path)
    df_ocgrt = df_ocgrt[df_ocgrt['Jurisdiction'] == 'NAT_TOTAL']
    df_Ox_country = df_ocgrt[df_ocgrt.CountryName== countries]
    df_Ox_country['Date'] = df_Ox_country['Date'].apply(lambda x: dt.datetime.strptime(str(x),'%Y%m%d'))
    df_Ox_country['Daily_cases'] = df_Ox_country['ConfirmedCases'].diff()
    
    ################## Merge tweet and OxCGRT
    df_Ox_country_tweets = df_Ox_country.merge(df_like_index,'left', on = 'Date')
    
    ########### Chart
    df_Ox_country_tweets.plot(x='Date',y=['like_index','retweet_index'],figsize=(10,4))
    ax2 = plt.twinx()
    sns.lineplot(x =df_Ox_country_tweets.Date, y = df_Ox_country_tweets.Daily_cases, ax = ax2,linewidth=.4,color='red')
    plt.savefig('Picture/Preliminary_plot_{}.png'.format(countries))    
    
    df_va =pd.read_csv(oxcgrt_va_path)
    df_va = df_va[df_va.CountryName == countries]
    df_va['Date'] = df_va['Date'].apply(lambda x: dt.datetime.strptime(str(x),'%Y%m%d'))
    df_Ox_country_tweets_va = df_Ox_country_tweets.merge(df_va,'left',on='Date')
    
    df_Ox_country_tweets_va.to_csv('Data/{}_for_checking.csv'.format(countries))
    
    with open('Data/columns.txt','w') as f:
        for column in df_Ox_country_tweets_va.columns.to_list():
            f.write(column + '\n')
    f.close()
    return df_Ox_country_tweets_va
```

```python
def pre_processing(db):
    
    flag_columns = ['C1', 'C2', 'C3', 'C4','C5','C6','C7', 'E1', 'H1', 'H6', 'H7', 'H8']
    social_media = ['like_index','retweet_index']
    covid_cases = ['ConfirmedCases', 'ConfirmedDeaths', 'Daily_cases']
    general_info = ['CountryName_x', 'CountryCode_x', 'Jurisdiction','Date']
    reference = ['StringencyIndex', 'StringencyIndexForDisplay', 'StringencyLegacyIndex', 'StringencyLegacyIndexForDisplay', 'GovernmentResponseIndex', 'GovernmentResponseIndexForDisplay', 'ContainmentHealthIndex', 'ContainmentHealthIndexForDisplay', 'EconomicSupportIndex', 'EconomicSupportIndexForDisplay']
    removed = ['M1_Wildcard', 'CountryName_y', 'CountryCode_y', 'RegionName','RegionCode']
    policy = ['V1_Vaccine Prioritisation (summary)', 'V2_Vaccine Availability (summary)', 'V3_Vaccine Financial Support (summary)']
    country = db.CountryName_x[0]
    
    db.update(db[policy+covid_cases].fillna(0))
    db.drop(removed+reference, axis=1, inplace=True)
    columns = db.columns
    
    for column in columns:
        if "_Flag" in str(column):
            db.drop(column, axis=1, inplace=True)
        elif not any(column in _list for _list in [social_media, covid_cases, general_info, reference, policy]):
            if str(column)[0:2] in flag_columns:
                policy.append(column)
                db[column] = np.select([db[column]==0, db[column]>0], [0, db[column]*2 + db[str(column)[0:2]+'_Flag']-1])
            elif str(column)[0:2] in ['V1', 'V2', 'V3']:
                db.drop(column, axis=1, inplace=True)
            else:
                policy.append(column)
        else:
            pass
    #db.dropna(subset=policy, inplace=True)
    db.to_csv('Data/Full_{}_2.0.csv'.format(country))
    return db
```

```python
datadir = os.path.join(os.getcwd(), "Data")
tweet_path = os.path.join(datadir, "Tweets")
countries = ['Brazil', 'India','United States','Mexico']
oxcgrt_path = 'Data/covid-policy-tracker-master/data/OxCGRT_latest.csv'
oxcgrt_va_path = 'Data/covid-policy-tracker-master/data/OxCGRT_vaccines_full.csv'
#tweet_file = ['Finalized_Brazil.csv','Finalized_India.csv','Finalized_United States.csv','Finalized_Mexico.csv']
```

```python
dbs = init_db(countries, tweet_path)

tweets_dbs = fetch_data(dbs, tweet_path, countries)
[US, Mexico, Brazil, India] = tweets_combination(tweet_path, tweets_dbs)

print("United States", len(US))
print("Mexico", len(Mexico))
print("Brazil", len(Brazil))
print("India", len(India))
```

```python
for country in countries:
    datafile = "Finalized_"+ country +".csv"
    full_db = pre_processing(source_combination(country, os.path.join(tweet_path, datafile), oxcgrt_path, oxcgrt_va_path))
    print(full_db)
```
